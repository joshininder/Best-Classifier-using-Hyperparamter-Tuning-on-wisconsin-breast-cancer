# Analysis Of Classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CSVs/Stats.csv", index_col = 'Classifier')
df.head(6)
```

```python
accuracy = df.iloc[:,0].values
precision = df.iloc[:, 1].values
f1 = df.iloc[:,2].values
recall = df.iloc[:,3].values
auc = df.iloc[:,4].values
```

```python
value = df.min()
value = list(int(i) for i in value[:4])
print(type(value))
print(value)
```

```python
%matplotlib inline

x = df.index
fig = plt.figure()
ax = fig.add_axes([1.5,1.5,1.5,1.5])
ax.plot(x,accuracy, label ='Accuracy')
ax.plot(x,precision, label = 'Precision')
ax.plot(x,recall,label = 'Recall')
ax.plot(x,f1, label = 'F1 Score')
ax.legend(loc = 'best')
plt.xticks(x)
plt.yticks(range(min(value),100,1))
plt.title('Accuracy, Precision, Recall, F1 Score Curves on Each Fold')
plt.show()
```
